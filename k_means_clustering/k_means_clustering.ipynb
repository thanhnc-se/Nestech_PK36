```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets._samples_generator import make_blobs
```

```python
# Tạo tập dữ liệu X gồm 500 item (n_samples), tập trung xung quanh 4 điểm (centers),
# giữ nguyên giá trị sau các lần chạy khác nhau (random_state=0), độ lệch chuẩn (cluster_std) = 0.5, khá nhỏ, dữ liệu chủ yếu sẽ tập chung xung quanh center
X, y = make_blobs(n_samples=500, centers=4, random_state=0, cluster_std=0.5)

# Danh sách màu: yellow, green, blue, magenta
colors = ["y", "g", "b", "m"]

# Tương tự như khi viết
# colors_scatter = []
# for i in y:
#     colors_scatter.append(colors[i])
colors_scatter = list(map(lambda i: colors[i], y))

plt.scatter(X[:, 0], X[:, 1], c=colors_scatter)
```

```python
def showScatterWithCenter(X, centers, pred_y):
    """
    Tô màu cho các điểm dữ liệu bằng kết quả thu được từ hàm predict
    sau đó đánh dấu các điểm center bằng màu đỏ
    """
    pred_colors = list(map(lambda x: colors[x], pred_y))
    plt.scatter(X[:, 0], X[:, 1], c=pred_colors)
    plt.scatter(centers[:, 0], centers[:, 1], s=300, c="red")
    plt.show()
```

```python
# Tạo kmeans model với số lượng cụm = 4, danh sách các center ban đầu ko thay đổi
kmeans = KMeans(n_clusters=4, random_state=0)
# Huấn luyện model với tập dữ liệu x
kmeans.fit(X)
# Dự đoán các item trong tập dữ liệu X thuộc cluster nào
kmeans.predict(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
pred_y = kmeans.predict(X)
centers = kmeans.cluster_centers_
centers
showScatterWithCenter(X, centers, pred_y)
```
