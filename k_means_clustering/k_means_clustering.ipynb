```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets._samples_generator import make_blobs
```

```python
# To tp d liu X gm 500 item (n_samples), tp trung xung quanh 4 im (centers),
# gi nguyn gi tr sau cc ln chy khc nhau (random_state=0),  lch chun (cluster_std) = 0.5, kh nh, d liu ch yu s tp chung xung quanh center
X, y = make_blobs(n_samples=500, centers=4, random_state=0, cluster_std=0.5)

# Danh sch mu: yellow, green, blue, magenta
colors = ["y", "g", "b", "m"]

# Tng t nh khi vit
# colors_scatter = []
# for i in y:
#     colors_scatter.append(colors[i])
colors_scatter = list(map(lambda i: colors[i], y))

plt.scatter(X[:, 0], X[:, 1], c=colors_scatter)
```

```python
def showScatterWithCenter(X, centers, pred_y):
    """
    T mu cho cc im d liu bng kt qu thu c t hm predict
    sau  nh du cc im center bng mu 
    """
    pred_colors = list(map(lambda x: colors[x], pred_y))
    plt.scatter(X[:, 0], X[:, 1], c=pred_colors)
    plt.scatter(centers[:, 0], centers[:, 1], s=300, c="red")
    plt.show()
```

```python
# To kmeans model vi s lng cm = 4, danh sch cc center ban u ko thay i
kmeans = KMeans(n_clusters=4, random_state=0)
# Hun luyn model vi tp d liu x
kmeans.fit(X)
# D on cc item trong tp d liu X thuc cluster no
kmeans.predict(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
pred_y = kmeans.predict(X)
centers = kmeans.cluster_centers_
centers
showScatterWithCenter(X, centers, pred_y)
```
